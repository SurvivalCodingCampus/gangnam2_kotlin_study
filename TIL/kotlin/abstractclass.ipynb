## 추상 클래스 (abstract class)

- 상속의 재료로 사용되는 클래스
- 상세 부분이 일부 미정인 클래스

### 추상 클래스(abstract class)와 추상 메소드(abstract method)

- 오버라이드 강제
- 추상클래스는 인스턴스화가 금지되어 있다

```kotlin
// 추상 클래스
abstract class Character(
    var name: String,
    var hp: Int,
) {
    fun run() = println("$name 이 도망갔다")

    // 추상 메서드
    abstract fun attack(slime: Slime)
}
```
